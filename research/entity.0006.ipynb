```python
nodes = pd.read_csv("dataset/nodes.csv")
edges = pd.read_csv("dataset/edges.csv")
```

```python
matrix = np.zeros((len(nodes), len(nodes)))
```

```python
n = len(nodes)
```

```python
matrix
```

```python

for _, row in edges.iterrows():
        matrix[row["source"], row["target"]] = 1
        matrix[row["target"], row["source"]] = 1
```

```python
communities = {}
for _, row in nodes.iterrows():
  if communities.get(row['groups']) == None:
    communities[row['groups']] = []
  communities[row['groups']].append(row['name'])
```

```python
k = []
for community in communities.keys():
  k.append(community)
k = np.array(k)
```

```python
k
```

```python
communities
```

```python
initial_population = []
for _ in range(100):
  individual = np.random.choice(k, size=n)
  initial_population.append(individual)
```

```python
def modularity(individual, matrix, edges):
  # Q = (1/2m) * Σ(Aij - (ki*kj)/(2m))δ(ci, cj)
  m = len(edges)
  # print(m)
  Q = 0
  k = [0] * n
  for i in range(n):
    k[i] = sum(matrix[i])
  for i in range(n):
    for j in range(n):
      if individual[i] == individual[j]:
        kikj = k[i]*k[j]
        ext = kikj/(2*m)
        Q+=matrix[i,j]-ext
  return Q/(2*m)
```

```python
selections = []
```

```python
matrix
```

```python
for individual in initial_population:
  score = modularity(individual, matrix, edges)
  selections.append((individual, score))
selections.sort(key=lambda x: x[1], reverse=True)
```

```python
selections
```

```python
best_selections = selections[:len(selections)//4]
```

```python
best_selections
```

```python
def mutation(chromosome):
  random_index = np.random.randint(0, len(chromosome))
  new_community = np.random.choice(k)
  chromosome[random_index] = new_community
  return chromosome
```

```python
def crossover(A, B):
  random_index = np.random.randint(0, n)
  community_selected = A[random_index]
  index_selected = []
  for i in range(n):
    if A[i] == community_selected:
      index_selected.append(i)
  for index in index_selected:
    B[index] = community_selected
  return B
```

```python
for i in range(len(best_selections)):
  if i != len(best_selections)-1:
    A = best_selections[i][0]
    B = best_selections[i+1][0]
    tmp_b = B.copy()
    B = crossover(A, B)
    A = crossover(tmp_b, A)
    best_selections[i+1] = (A, modularity(A, matrix, edges))
    best_selections[i] = (B, modularity(B, matrix, edges))
  else:
    A = best_selections[i][0]
    B = best_selections[0][0]
    tmp_b = B.copy()
    B = crossover(A, B)
    A = crossover(tmp_b, A)
    best_selections[0] = (A, modularity(A, matrix, edges))
    best_selections[i] = (B, modularity(B, matrix, edges))
    
best_selections.sort(key=lambda x: x[1], reverse=True)
```

```python
best_selections
```

```python
# mutation 
mutation_gen = best_selections.copy()
for i in range(len(mutation_gen)):
  chromosome = mutation_gen[i][0]
  mutated_chromosome = mutation(chromosome)
  mutation_gen[i] = (mutated_chromosome, modularity(mutated_chromosome, matrix, edges))
```

```python
mutation_gen
```

```python
# local search

T = 800000
x = 0.99
l = 10
step = 0.02

initial_population = []
for _ in range(1000):
  individual = np.random.choice(k , size=n)
  initial_population.append(individual)

res = (0, [])
anneal_time = 1

while T > 1:
  g1_index = np.random.randint(0, len(initial_population))
  g2_index = np.random.randint(0, len(initial_population))
  g1 = initial_population[g1_index]
  g2 = initial_population[g2_index]
  limit = 0.0
  if limit >= 1.0:
    T = T * x
    continue

  for i in range(l):
    g2_temp = g2.copy()
    G1 = crossover(g1, g2)
    G2 = crossover(g2_temp, g1)
    mutation_rate = random.uniform(0, 1)
    if mutation_rate < 0.3:
      G1 = mutation(G1)
      G1 = mutation(G1)
    else:
      G1 = mutation(G1)
      G2 = mutation(G2)
    p = modularity(G1, matrix, edges) - modularity(G2, matrix, edges)
    r = random.uniform(0, 1)
    if p > limit or np.exp(p/T) >  r:
      g1 = G1
      limit += step
  T = T * x
  q = modularity(g1, matrix, edges)
  if T < 1:
    print(f"Anneal {anneal_time}: Modularity = {q}, Gene = {g1}")
    if q-res[0] >= 0.2:
      print("Found")
      res = (q, g1)
      break
    else: 
      anneal_time += 1
      T = 800000

print(res)
```
